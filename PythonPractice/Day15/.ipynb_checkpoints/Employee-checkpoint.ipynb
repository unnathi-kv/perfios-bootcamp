```python
from tkinter import * 
from tkinter import ttk
import tkinter as tk
import datetime
import time
import tkinter.messagebox
import sqlite3
```

```python
class EmployeeInfo():
    
    db_name='employee.db'
    
    def __init__(self,root):
        self.root=root
        self.root.title("IVE Portal")
        
        
         #~~~~~~~~~~~~~~~~~~Display from database~~~~~~~~~~~~~~~~~~~~~~
        
        def run_query(query,parameters=()):
            with sqlite3.connect(self.db_name) as conn:
                cursor = conn.cursor()
                query_result = cursor.execute(query,parameters)
                conn.commit()
            return query_result
        
        def viewing_records():
#             print(self.UNAME_LBL.get())
            records = self.tree.get_children()
            for element in records:
                self.tree.delete(element)
            query='select * from info'
            db_table=run_query(query)
            for data in db_table:
                self.tree.insert('',tk.END,text=data[0],values=data[1:])
    
       
        
        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Add Data~~~~~~~~~~~~~~~~~~~~~~~~~~
        
        self.message=Label(text='',fg='red')
        self.message.grid(row=8,column=15)
        
        #Adding records 
        def validation():
            return len(self.UID_TXT.get())!= 0 and len(self.NAME_TXT.get()) != 0 and len(self.DEPT_TXT.get()) != 0 
        
        
        def add_record():
#             print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
            if validation():
                query='insert into info values(?,?,?)'
                parameters=(self.UID_TXT.get(),
                            self.NAME_TXT.get(),
                            self.DEPT_TXT.get())

                run_query(query,parameters)
                self.message['text']='Record {} is added'.format(
                    self.UID_TXT.get()
                )
                viewing_records()

            else:
                self.message['text']='Fields are not completed'
            
        def delete_data():
            current_item = self.tree.focus()
            values = self.tree.item(current_item)
            selected_item = self.tree.selection()[0]
            self.tree.delete(selected_item)
            query = 'delete from info where emp_id=?'
            para = (values['text'],)
#             print(para)
            run_query(query,para) 
            self.message['fg'] = 'green'
            self.message['text'] = 'Deleted record with EMPLOYEE ID = {}'.format(para[0])

        
        
    
        #............Logo and Title...........
        
        self.photo=PhotoImage(file='emplogo (1).png')
        self.label=Label(image=self.photo)
        self.label.grid(row=0,column=0)
        self.label1=Label(font=('times new roman',15,'bold'),bd=12,relief=GROOVE,text='IVE Employees', bg='#2B435D' ,
                          fg='#FBFFFF')
        self.label1.grid(row=1,column=0) 
        
        
        #.........LABELS and ENTRY BOX...........
            
            
        self.frame=LabelFrame(self.root,text="Add New Record",fg='#2B435D',font=("times new roman",15,"bold"))
        self.frame.grid(row=0,column=1,padx=5,pady=10)
        
        self.UID_LBL=Label(self.frame,text="Employee ID: ",fg="#2B435D",font=("times new roman",12,"bold"))
        self.UID_LBL.grid(row=1,column=10,padx=20,pady=5)
        self.UID_TXT=Entry(self.frame,width=15,font="arial 15",bd=7,textvariable="uname",relief=RIDGE)
        self.UID_TXT.grid(row=1,column=15,pady=5,padx=10)
        
        self.NAME_LBL=Label(self.frame,text="Employee Name: ",fg="#2B435D",font=("times new roman",12,"bold"))
        self.NAME_LBL.grid(row=2,column=10,padx=20,pady=5)
        self.NAME_TXT=Entry(self.frame,width=15,font="arial 15",bd=7,textvariable="name",relief=RIDGE)
        self.NAME_TXT.grid(row=2,column=15,pady=5,padx=10)
        
        self.DEPT_LBL=Label(self.frame,text="Department : ",fg="#2B435D",font=("times new roman",12,"bold"))
        self.DEPT_LBL.grid(row=3,column=10,padx=20,pady=5)
        self.DEPT_TXT=Entry(self.frame,width=15,font="arial 15",bd=7,textvariable="age",relief=RIDGE)
        self.DEPT_TXT.grid(row=3,column=15,pady=5,padx=10)
        
        self.add_btn=Button(self.frame,text="Add",width=10,bd=7,font="arial 12 bold",bg="#2B435D",fg='#FBFFFF',command=add_record)
        self.add_btn.grid(row=7,column=15,padx=15,pady=10) 
        
        
        
        
       #..........Database display box........
        
        self.tree=ttk.Treeview(height=10,column=['',''])
        self.tree.grid(row=0,column=3,columnspan=2)
        self.tree.heading('#0',text='Employee ID')
        self.tree.column('#0',width=100)
        
        self.tree.heading('#1',text='Employee Name')
        self.tree.column('#1',width=100)
        
        self.tree.heading('#2',text='Department')
        self.tree.column('#2',width=150)
        
        self.add_btn=Button(text="Delete",width=10,bd=7,font="arial 12 bold",bg="#2B435D",fg='#FBFFFF',command=delete_data)
        self.add_btn.grid(row=1,column=3)
        

        #..............Display date and time.............
        
        def tick():
            d=datetime.datetime.now()
            Today='{:%B %d, %Y}'.format(d)
            mytime=time.strftime('%I: %M: %S %p')
            self.lblInfo.config(text=("Time:" + mytime + '\t' +Today))
            self.lblInfo.after(200,tick)

        self.lblInfo=Label(font=('arial',15,'bold'),fg='#2B435D',justify=LEFT)
        self.lblInfo.grid(row=0,column=5)
    
        tick()
        
         
        #...............Add Menu Bar..............
        Chooser = Menu()
        itemone = Menu()

        
        
        itemone.add_command(label='Add Record')
        itemone.add_command(label='Edit Record')
        itemone.add_command(label='Delete Record')
        itemone.add_command(label='Help')
        itemone.add_command(label='Exit')



        Chooser.add_cascade(label='File',menu=itemone)
        Chooser.add_cascade(label='Help')
        Chooser.add_cascade(label='Exit')
        root.config(menu=Chooser)
        
        viewing_records()
        
        
        
if __name__=='__main__':
    root=Tk()
    root.geometry('1500x735+150+200')
    application =EmployeeInfo(root)
    root.mainloop()
```
