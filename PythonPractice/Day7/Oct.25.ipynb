```python
def starts_with(pattern,word):
    p=r'^{}'.format(pattern)
    r = re.search(p,word)
    if r:
        return True
    else:
        return False

starts_with('my','my name is dilip')
```

```python
def ends_with(pattern,word):
    p=fr'{pattern}$'
    r = re.search(p,word)

    return re.search(p,word) != None
   
ends_with("02","56156165102 2132123102")
```

```python
class Average:
    """
    A sequential avg calculator
    Useage:
    -> a = Average(10)
    -> a.add(20)
    15
    -> a.add(30)
    20
    """
    lst=[]
    def __init__(self,v):
        self.lst.append(v)
     
    def avg(self):
        print(f"""The given inputs are : {self.lst} \nThe average of the values is : {sum(self.lst)/len(self.lst)}""")
        
    def add(self,v):
        self.lst.append(v)
        self.avg()
    
```

```python
a = Average(10)
```

```python
a.add(20)
```

```python
a.add(30)
```
