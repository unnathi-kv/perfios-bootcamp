```python
import matplotlib.pyplot as plt
# markers - o , * , + , . etc
y=np.array([3,8,1,15])
plt.plot(y,marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
# line style - dashed,dotted
y=np.array([3,8,1,15])
plt.plot(y,marker='o',linestyle='dashed')
plt.show()
```

```python
# line color 
y=np.array([3,8,1,15])
plt.plot(y,marker='.',linestyle='dashed',c='hotpink')
plt.show()
```

```python
# graph labels and title for graphs
y=np.array([3,8,1,15])
plt.plot(y,marker='.',linestyle='dashed',c='hotpink')
plt.xlabel("Timeeeeeeee")
plt.ylabel("Speeeeeeeeed")
plt.title("Accelraaaaaaaaaaaation")
plt.show()
```

```python
# fonts for title and labels using fontdict attribute and position of title using loc attribute
y=np.array([3,8,1,15])
plt.plot(y,marker='.',linestyle='dashed',c='black')
font1 = {'family':'serif','color':'hotpink','size':15}
plt.xlabel("Time",fontdict={'size':11,'color':'purple'})
plt.ylabel("Speed",fontdict={'size':11,'color':'purple'})
plt.title("Acceleration",fontdict=font1,loc='left')
plt.grid(axis='y',color='green',linewidth = 1.5)
# plt.grid(axis='x',color='red',linestyle='dashed',linewidth=1.5)
plt.show()
```

```python
# subplot(m,n,p) - subplot( m , n , p ) divides the current figure into an 
# m -by- n grid and creates axes in the position specified by p i.e m -rows n - columns p -position of graph
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1) # 2- rows , 2- columns
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([100, 200, 300, 400])

plt.subplot(2,2 ,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([100, 200, 300, 400])

plt.subplot(2,2 ,4)
plt.plot(x,y)

plt.show()
```

```python
# scatter plot is used to  show how much one variable is affeced by another
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
# print(len(x))
# print(len(y))
plt.scatter(x,y,marker ="^",alpha=0.5,c="hotpink",cmap="viridis") # x and y should have same length of arrays 
plt.colorbar() # to show cmap
plt.show()
```

```python
X = ['Day 1','Day 2','Day 3','Day 4']
y = [10,20,30,40]
x = [20,30,40,50]

X_axis = np.arange(len(X))
plt.bar(X_axis ,x,0.4,label="BMW")
plt.bar(X_axis + 0.2 , y, 0.4 , label="Audi")
plt.xlabel("Days")
plt.ylabel("Distance")
# plt.title("Number of Students in each group")
plt.legend()
plt.show()
```

```python
bmw_x = np.array([0,1,2,3,4])
bmw_y = np.array([55,68,87,69,78])

audi_x = np.array(bmw_x)
audi_y = np.array([65,28,63,22,12])

plt.bar(bmw_x - 0.2,bmw_y, 0.4, label = "BMW")
plt.bar(audi_x + 0.2,audi_y,0.4, label = "Audi")


plt.legend()
plt.show()
```
