### Latest News using BeautifulSoup:- 

```python
import requests
status = requests.get("https://www.google.com/search?q=gold+value+today&oq=gold+va&aqs=chrome.1.0i512j0i131i433i512j69i57j0i512l7.2453j0j7&sourceid=chrome&ie=UTF-8")
status.text
```

```python
try:
    f4=open("goldvalue.txt",'w')
    f4.write(status.text)
except Exception as e:
    print(repr(e))
finally:
    f4.close()
```

```python
from bs4 import BeautifulSoup
import requests
# url="https://www.google.com/search?q=gold+value+today&oq=gold+va&aqs=chrome.1.0i512j0i131i433i512j69i57j0i512l7.2453j0j7&sourceid=chrome&ie=UTF-8"
url="https://timesofindia.indiatimes.com/?from=mdr"
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
# print(soup)
data=soup.find_all('figcaption')
for item in data:
    print(item.text)
```
